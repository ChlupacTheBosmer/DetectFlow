## Example of using detectflow class

First we will use the Sampler class to create a sample image with bounding boxes. Then we will use the Inspector class to display the image with bounding boxes. Note that the image is completely random and the bounding boxes are also random.

Second we will print the results of the bounding boxes. The variable boxes which is of type detectflow.DetectionBoxes (class instance) has two attributes: xyxy and xywh. The xyxy attribute is the bounding boxes in the format of (x1, y1, x2, y2) and the xywh attribute is the bounding boxes in the format of (x, y, w, h). We will print both of them.

Finally, we create a short function that simulates a prediction. The function will create a number of frames with random images and bounding boxes. The function will return a list of images and a list of bounding boxes. The function also has an optional parameter inspect which will display the images with bounding boxes if set to True.

The scripts demonstrate how to use classes and functions from detectflow. 

```python
from detectflow import Sampler, Inspector

image, boxes = Sampler.create_sample_image_with_bboxes(as_detection_boxes=True)

Inspector.display_frames_with_boxes([image], [boxes])
```

```python
print(boxes)
print(boxes.xyxy)
print(boxes.xywh)
```

```python
from detectflow import Sampler, Inspector

def sim_prediction(frames_num: int = 10, inspect: bool = False):
    list_of_images = []
    list_of_boxes = []
    list_of_frames = list(range(0, frames_num))
    for frame in list_of_frames:
        
        print(f"Frame number: {frame}")
        
        image, boxes = Sampler.create_sample_image_with_bboxes(as_detection_boxes=True)

        if inspect:
            Inspector.display_frames_with_boxes([image], [boxes])
        else:
            print("No insepction required")
        
        list_of_images.append(image)
        list_of_boxes.append(boxes)
        
    return list_of_images, list_of_boxes
```

```python
images, boxes = sim_prediction(20)

Inspector.display_frames_with_boxes(images, boxes)
```
